Function:

Python functions are reusable blocks of code used to perform a specific task. They help organize programs into smaller sections and execute the same logic whenever needed by calling the function. We use the def keyword to define a function.

```python
def name():
    print("Michael Jordan")

name()
```

```python
def add(a, b):
    return a + b

result = add(3, 4)
print(result) 
```

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

gen = count_up_to(3)

print(next(gen)) 
print(next(gen)) 
print(next(gen)) 
```

```python
def squares(n):
    for i in range(n):
        yield i * i

for x in squares(5):
    print(x)
```

```python
def square(x):
    return x ** 2

numbers = [1, 2, 3, 4, 5]

result = list(map(square, numbers))

print(result)
```

```python
square = lambda x: x * x

print(square(5))
```

```python
from functools import reduce
```

```python
numbers = [1, 2, 3, 4, 5]

result = reduce(lambda x, y: x + y, numbers)

print(result)
```

```python
numbers = [1, 2, 3, 4, 5, 6]

def is_even(n):
    return n % 2 == 0

result = list(filter(is_even, numbers))
print(result)
```
