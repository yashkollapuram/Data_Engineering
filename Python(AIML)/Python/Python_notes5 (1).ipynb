File Operations:
File operations in Python are used to read, write, append files. Python provides built-in functions to handle files efficiently.

write mode:
write mode is used to open a file for writing. It is specified with the mode 'w'.

Characteristics of 'w' mode:

1) Creates the file if it does not exist.
2) Overwrites (erases) all existing content if the file already exists.
3) Allows only writing operations.

```python
file = open("example.txt", "w")
file.write("Welcome to the NBA finals")
file.close()
```

```python
file = open("example.txt", "w")
file.write("It's San Antonio Spurs vs New York Knicks\n")
file.write("Here at the Frost Bank Center in San Antonio\n")
file.close()
```

read mode:
read mode is used to open a file for reading. It is specified with the mode 'r.

Characteristics of 'r' mode: 

1) Opens an existing file for reading.
2) Raises a "FileNotFoundError" if the file does not exist.
3) Does not allow writing to the file.

```python
file = open("example.txt", "r")
content = file.read()
print(content)
file.close()
```

append mode:
append mode is used to add data to the end of an existing file without deleting its current contents. It is specified with the mode 'a'.

Characteristics of 'a' mode:

1) Opens a file for writing at the end of the file.
2) Creates the file if it does not exist.
3) Preserves existing content.
4) New data is added after the existing content.

```python
file = open("example.txt", "a")
file.write("New York Knicks win game 1 here in San Antonio\n")
file.close()
```

```python
with open("example.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
with open("example.txt", "a") as file:
    file.write("\nIt was a close game but the Knicks steal game 1")
```

r+ mode:
'r+' opens a file for both reading and writing.

```python
with open("example.txt", "r+") as file:
    content = file.read()
    print(content)
    file.write("\nNow we look forward to game 2")
```

seek():
seek() is used to move the file pointer (cursor) to a specific position inside a file.

```python
with open("example.txt", "r") as file:
    file.seek(3)
    print(file.read())
```

Errors and exceptions:
Exceptions are runtime errors that interrupt the normal flow of a program. Python provides the try, except, else and finally blocks to handle exceptions gracefully.

Try:
Code that might raise an exception is placed inside a try block.

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num
```

If an error occurs, Python immediately stops executing the rest of the try block and looks for a matching except block.

except Block:
The except block catches and handles exceptions.

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ZeroDivisionError:
    print("Cannot divide by zero.")
except ValueError:
    print("Please enter a valid integer.")
```

else Block:
The else block runs only if no exception occurs in the try block.

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ZeroDivisionError:
    print("Cannot divide by zero.")
except ValueError:
    print("Invalid input.")
else:
    print("Result =", result)
```

finally Block:

The finally block executes whether an exception occurs or not. It is commonly used for cleanup operations such as closing files, database connections or releasing resources.

The finally block runs even though an exception occurrs.

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ZeroDivisionError:
    print("Cannot divide by zero.")
finally:
    print("Execution completed.")
```
