1. Encapsulation:

Encapsulation means bundling data and methods together and controlling access to internal details.

i) Composition:

Composition means building complex objects using other objects.

```python
class Engine:
    def start(self):
        print("Engine started")

class Car:
    def __init__(self):
        self.engine = Engine()  # Composition

    def drive(self):
        self.engine.start()
        print("Car moving")

car = Car()
car.drive()
```

ii) Dynamic Extension:

Python objects can be modified at runtime.

```python
class Student:
    pass

s = Student()

s.name = "John"
s.age = 20

print(s.name)
```

2. Polymorphism:

Polymorphism means one interface, many implementations.

```python
class Dog:
    def speak(self):
        return "Woof"

class Cat:
    def speak(self):
        return "Meow"

animals = [Dog(), Cat()]

for animal in animals:
    print(animal.speak())
```

3. Duck Typing:

Python focuses on behavior rather than type.

```python
class Dog:
    def speak(self):
        print("Woof")

class Robot:
    def speak(self):
        print("Robot voice")

def make_sound(obj):
    obj.speak()

make_sound(Dog())
make_sound(Robot())
```

4. Class Variable: Shared by all instances.

Instance Variable: Unique to each object.

```python
class Student:
    school = "Green High School"  # class variable

    def __init__(self, name):
        self.name = name  # instance variable

s1 = Student("Alice")
s2 = Student("Bob")

print(s1.school)  # Green High School
print(s2.school)  # Green High School
```

5. Instance Method:

Works on object instances.

```python
class Student:
    def __init__(self, name):
        self.name = name

    def display(self):
        print(self.name)

s = Student("Chris")
s.display()
```

6. Class Method:

Works on the class itself.

```python
class Student:
    count = 0

    def __init__(self):
        Student.count += 1

    @classmethod
    def get_count(cls):
        return cls.count

s1 = Student()
s2 = Student()

print(Student.get_count())
```

7. Static Method:
Independent utility function inside a class.

```python
class Calculator:
    @staticmethod
    def add(a, b):
        return a + b

# Call using the class
result1 = Calculator.add(3, 5)
print(result1)  # 8

# Call using an instance
calc = Calculator()
result2 = calc.add(10, 20)
print(result2)  # 30
```

8. Decorators:

Decorators modify the behavior of functions or methods.

```python
def decorator(func):

    def wrapper():
        print("Before")
        func()
        print("After")

    return wrapper

@decorator
def greet():
    print("Hello")

greet()
```

9. Dataclass:

A dataclass automatically generates common methods like:

__init__
__repr__

```python
from dataclasses import dataclass

@dataclass
class Employee:
    name: str
    salary: float

e = Employee("Alice", 50000)

print(e)
```

Single Responsibility Principle (SRP):
A class should have one reason to change. A class should focus on one responsibility or one area of concern.

```python
class Employee:
    pass


class EmployeeRepository:
    def save(self, employee):
        pass


class ReportGenerator:
    def generate(self, employee):
        pass


class EmailService:
    def send(self, employee):
        pass
```
