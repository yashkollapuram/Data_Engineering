What is OOP?

Object-Oriented Programming (OOP) is a programming paradigm that organizes code using objects and classes.

Class → Blueprint/template.
Object → Instance created from a class.

```python
class Car:
    pass

car1 = Car()   # object
car2 = Car()   # object
```

1. Class:
A class defines attributes (data) and methods (functions) that objects will have.

```python
class Student:
    school = "ABC School"  # class attribute
```

2. Object:
An object is an instance of a class.

```python
class Student:
    pass

s1 = Student()
s2 = Student()

print(type(s1))
```

3. Constructor:
A constructor is a special method automatically called when an object is created.

In Python, the constructor is typically implemented using __init__().

```python
class Student:
    def __init__(self):
        print("Object created")

s1 = Student()
```

4. __init__:

__init__ initializes object attributes.

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

s1 = Student("John", 20)

print(s1.name)
print(s1.age)
```

self:

self refers to the current object.

```python
class Demo:
    def show(self):
        print("Inside object")

d = Demo()
d.show()
```

Python internally does:

```python
Demo.show(d)
```

5. Methods:
Methods are functions defined inside a class.

```python
class Student:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print(f"Hello, I am {self.name}")

s1 = Student("Chris")
s1.greet()
```

6. *args:

*args allows a function to accept a variable number of positional arguments.

```python
def add(*args):
    print(args)
    return sum(args)

print(add(1, 2, 3, 4))
```

Using inside a class:

```python
class Numbers:
    def total(self, *args):
        return sum(args)

n = Numbers()
print(n.total(10, 20, 30))
```

7. **kwargs:

**kwargs allows a function to accept a variable number of keyword arguments.

```python
def show(**kwargs):
    print(kwargs)

show(name="John", age=20)
```

In a class:

```python
class Person:
    def __init__(self, **kwargs):
        self.info = kwargs

p = Person(name="Alice", age=25)

print(p.info)
```

8. __str__:

__str__ defines the readable string representation of an object.

```python
class Student:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Student: {self.name}"

s = Student("John")

print(s)
```

9. __repr__:

__repr__ provides the official/debug representation of an object.

```python
class Student:
    def __init__(self, name):
        self.name = name

    def __repr__(self):
        return f"Student('{self.name}')"

s = Student("John")

print(repr(s))
```

10. __dict__:

__dict__ stores an object's writable attributes in a dictionary.

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

s = Student("John", 20)

print(s.__dict__)
```

11. Abstraction:

Abstraction means hiding implementation details and exposing only essential functionality. Python provides abstraction through the abc module.

Benefits:

1) Hides complexity
2) Enforces a common interface
3) Improves maintainability

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass
```

```python
Any subclass must implement start().
```

```python
class Car(Vehicle):

    def start(self):
        print("Car started")

c = Car()
c.start()
```

12. Inheritance:

Inheritance allows one class to acquire properties and methods of another class.

```python
class Person:

    def introduce(self):
        print("I am a person")

class Student(Person):

    def study(self):
        print("I am studying")

s = Student()

s.introduce()   # inherited from Person
s.study()       # defined in Student
```
