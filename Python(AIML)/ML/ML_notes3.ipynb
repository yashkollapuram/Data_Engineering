1) Classification: A broad category of supervised learning tasks. The primary objective is to map input data into distinct, discrete labels (e.g., True/False, Cat/Dog, 0/1).

2) Logistic Regression: A specific machine learning algorithm. It is the most common tool used to solve binary classification problems. It’s continuation of linear regression.

3) Sigmoid Function: A sigmoid function is a mathematical function with a characteristic "S"-shaped curve. It is most used in machine learning and statistics to map any real-valued number into a probability bounded between 0 and 1. It’s a binary classification.

4) Threshold Values: Threshold value classification is a method of categorizing data based on whether they fall above, below or within specified threshold values (cutoff points).

Ex: >0.4 - cancer, <0.4 - non-cancer

5) Softmax Function: It converts a vector of real numbers (logits) into a probability distribution whose values sum to 1. Total sum of probabilities of all classes is 1 and the highest value is the class.

In a multiclass classification model that uses softmax, the predicted class is usually the one with the highest softmax probability.

For example, if the softmax output is:
0.65, 0.25, 0.10

and these correspond to classes:

Class A: 0.65
Class B: 0.25
Class C: 0.10

then the model predicts Class A because it has the highest probability (0.65).

Evaluation of Classification model:

Metrics: Accuracy, Recall, Precision, F1 score, specifity, AUC, RUC.

Accuracy: Percentage of total predictions that are correct.
Formula: (TP + TN) / (TP + TN + FP + FN)
Ex: If 90 out of 100 predictions are correct, Accuracy = 90%

Recall: Measures how many actual positive cases are correctly identified.
Formula: TP / (TP + FN)
Ex: Out of 20 actual positives, model finds 16 → Recall = 16/20 = 0.80

Precision: Measures how many predicted positives are actually positive.
Formula: TP / (TP + FP)
Ex: Model predicts 25 positives, 20 are correct → Precision = 20/25 = 0.80

F1 score: Harmonic mean of Precision and Recall. Useful when classes are imbalanced.

Formula: 2 × (Precision × Recall) / (Precision + Recall)

Ex: Precision = 0.8, Recall = 0.8 → F1 = 0.8

Specifity: Measures how many actual negative cases are correctly identified.
Formula: TN / (TN + FP)
Ex: Out of 50 negatives, model correctly identifies 45 → Specificity = 45/50 = 0.90

ROC curve: Graph showing True Positive Rate (Recall) vs False Positive Rate at different thresholds. Range of values is between 0 and 1.

AUC: Measures the model's ability to distinguish between positive and negative classes. It helps choose the best model amongst the models. The best model is the one which has the max area.

Precision–Recall (PR) Curve:

A Precision–Recall Curve is a graph that shows the trade-off between Precision and Recall at different classification thresholds.
