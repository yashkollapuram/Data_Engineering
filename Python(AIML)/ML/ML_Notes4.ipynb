Decision Tree: 

A decision tree is a tree-shaped model used to make decisions or predictions by splitting data into smaller groups based on a series of questions or conditions. The goal is to achieve a pure node. It is a model that can be adapted for either classification or regression.

In a decision tree, the main components are:

i) Parent node: A node that splits into one or more child nodes. It represents a decision, test or condition.
ii) Child node: A node that comes directly from a parent node. A child node may itself become a parent if it further splits.
iii) Leaf node (terminal node): The final node of a branch. It has no children and represents the final decision, class or predicted value.

           Is it raining?          ← Parent node (root)
             /      \
          Yes        No            ← Child nodes
          /           \
 Take umbrella   Carry sunglasses  ← Leaf nodes

Recursive binary splitting:

Recursive binary splitting is the most common method used to build a decision tree. It is called a greedy approach because at each step it chooses the best split available at that moment, without considering future impact/splits. When there's too much variance in the data then the model becomes complex.

The algorithm chooses the locally best split at each step and does not revisit earlier decisions. Therefore, it may not produce the globally optimal tree.

Key point:
Recursive binary splitting = repeatedly choosing the best two-way split at each node until a stopping rule is met. This is the standard tree-growing procedure used in many decision tree algorithms.

Tree pruning:

Tree Pruning is the process of reducing the size of a decision tree to avoid overfitting and improve its performance on new data. It is used to reduce complexity in the model.

i) Post-Pruning:

It is also known as backward pruning. It grows the full tree first, then remove branches that do not improve performance.

ii) Pre-pruning:

It is also known as forward pruning. Stop growing the decision tree early based on certain rules (e.g., maximum depth or minimum samples per node).

Pre-pruning prevents overgrowth by stopping tree construction early, while post-pruning removes unnecessary branches after the full tree is built. Pre-pruning is computationally cheaper, whereas post-pruning often achieves better accuracy.

Entropy:

Entropy measures the impurity or randomness of data within a node. The goal of the algorithm is to split data so that leaf nodes are completely homogeneous (pure). It's the difference between the entropies before and after the split.

Low Entropy (0): The data is perfectly pure and belongs to a single class (e.g., all samples are "Yes" or all are "No").

High Entropy (1.0 for binary classes): The data is highly disordered and evenly split across multiple classes, making it impossible to predict the outcome.

Information Gain:

Information gain measures the reduction in entropy (uncertainty) after splitting a dataset based on a specific attribute or feature.

High information gain = the split makes groups much more “pure”.
Low information gain = split doesn’t help much.

Gini impurity:

Gini impurity measures how often you would be wrong if you randomly labeled an item based on class proportions. 

0 = perfectly pure (only one class)
Higher = more mixed

The value of ginni index which is less will be used as the tree node for splitting.

Different Algorithms For Decision Tree:

i) ID3
ii) C4.5
iii) CART

1. ID3 (Iterative Dichotomiser 3):

Idea: Builds a tree using Information Gain (Entropy reduction).

i) Chooses the feature that gives the highest information gain
ii) Handles categorical features only
iii) No pruning (can overfit easily)
iv) Doesn’t handle missing values well

2. C4.5 (Improved version of ID3):

Idea: Fixes limitations of ID3.

Key improvements:

Uses Gain Ratio (fixes bias toward many-valued features)
Can handle continuous (numeric) features
Can handle missing values
Uses pruning to reduce overfitting
Converts tree into if-then rules

3. CART (Classification and Regression Trees)

Idea: Most widely used modern tree method.

Key features:

Uses Gini Impurity (for classification)
Can also do regression trees (predict numbers)
Produces binary splits only (always two branches)
Strong pruning strategy
Handles numeric and categorical data

Advantages of Decision Trees:

1. Easy to understand and interpret:
They look like flowcharts, so even non-technical users can follow the decision rules.

2. Requires little data preparation:
No need for feature scaling or normalization, and they can handle both numerical and categorical data.

3. Handles nonlinear relationships well:
They can capture complex patterns without requiring transformation of inputs.

4. Works for classification and regression:
The same structure can be used for predicting categories or continuous values.

5. Feature importance is visible:
You can see which variables are most influential in making predictions.

6. Fast to train and predict (for small/medium trees):
Especially compared to more complex models like neural networks.

Disadvantages of Decision Trees:

1. Overfitting (major issue):
Trees can become too complex and memorize training data instead of generalizing well.

2. Unstable (high variance):
Small changes in data can produce a completely different tree structure.

3. Greedy algorithms are not optimal:
They make locally optimal splits, which may not lead to the best overall tree.

4. Biased toward dominant classes:
If the dataset is imbalanced, the tree may favor the majority class.

5. Poor performance on smooth relationships:
They approximate with step-like rules, which can be less accurate for continuous trends.

6. Can become complex and hard to interpret if deep:
Although small trees are simple, large ones become messy and less useful.

Cross-validation (CV):

Cross-validation is a statistical resampling technique used to evaluate how well a machine learning model generalizes to unseen data while preventing overfitting.

i) Hold out method:

The hold-out method is the simplest way to evaluate a model by splitting the dataset into separate parts.
Ex: 80% training / 20% testing

ii) K-Fold Cross Validation:

K-fold cross-validation is a model evaluation technique used to check how well a machine learning model performs on unseen data by using multiple train-test splits.

Example (K = 5)

Data is split into 5 folds: A, B, C, D, E

Round 1: Train on B, C, D, E → Test on A
Round 2: Train on A, C, D, E → Test on B
Round 3: Train on A, B, D, E → Test on C
Round 4: Train on A, B, C, E → Test on D
Round 5: Train on A, B, C, D → Test on E

Final score = average of all 5 test results

iii) Leave-One-Out Cross Validation (LOOCV):

LOOCV is the most extreme form of cross-validation, where each data point is tested individually, giving a very thorough but computationally heavy evaluation.

iv) Stratified K-Fold Cross Validation:

It is an improved version of k-fold cross-validation used mainly for classification problems, especially when classes are imbalanced.

So if your dataset has:

70% Class A
30% Class B

Then each fold will also roughly have 70% A and 30% B.

Hyper Parameters:
Hyperparameters are the settings or configurations of a machine learning model that are set before training begins and are not learned from the data.

Ensemble methods combine multiple models (called weak learners) to create a stronger and more accurate model.

Main types:

Bagging
Boosting

1. Baaging:

Train many models independently in parallel on different random samples of the data, then combine their outputs.

Random Forest:

A special type of bagging using decision trees, where randomness is added in feature selection.

Key idea:

At each split in a tree:
Only a random subset of features is considered

2. Boosting:

Train models sequentially, where each new model tries to fix errors made by previous models.
