Boosting is an ensemble machine learning technique that combines multiple weak learners (models that perform slightly better than random guessing) to create a strong learner with higher predictive accuracy.

1. AdaBoost (Adaptive Boosting):

Adaptive Boosting (AdaBoost) is a machine learning method that combines many simple models, where each new model focuses more on the mistakes made by the previous ones, to improve overall accuracy.

Ex: Each model “votes” on whether an email is spam or not:

Model 1 vote (weak but general)
Model 2 vote (corrects earlier mistakes)
Model 3 vote (refines further)

The final decision is a weighted majority vote, where better models have more influence.

Gradient Boosting:

Gradient Boosting is a machine learning technique where models are built sequentially, and each new model tries to correct the errors (residuals) made by the previous models using a method inspired by gradient descent.

Gradient Boosting builds models that learn from the residual errors of previous models and add corrections step by step.

Better handling of variance & error rates, less computation cost.

Ex: 

Step 1: First model
Model 1 predicts:
Actual price: $300k
Predicted: $250k
Error (residual): +$50k

Step 2: Second model
Model 2 is trained to predict the error ($50k) made by Model 1.
It learns patterns like:
“houses near schools are underpriced”

Step 3: Combine models

Final prediction =
Model 1 prediction + Model 2 correction

So:

$250k + $50k = $300k (better prediction)

Step 4: Repeat
More models keep learning remaining errors step by step.
Each one improves the overall prediction.

XG Boost:

XGBoost (Extreme Gradient Boosting) is a highly optimized and scalable version of Gradient Boosting that builds decision trees sequentially to reduce errors but does it faster, more accurately and with better control over overfitting.

XGBoost is an improved gradient boosting algorithm that uses advanced optimizations + regularization to make predictions more accurate and efficient.

Ex:
Predicting house prices:

First tree predicts roughly:
$250k (true = $300k)
Second tree learns:
“Add +$30k for location factor”
Third tree learns:
“Add +$20k for size factor”

Final result:
$250k + $30k + $20k = $300k

XGBoost ensures:

i) No over-correction
ii) No overly complex trees
iii) Faster computation

Catboost:
CatBoost is a gradient boosting method that works with categorical data.

Key features of CatBoost:
1. Handles categorical data automatically
2. Reduces overfitting

Ex:
Predicting whether a customer will buy a product:

City = “Mumbai”
Product category = “Electronics”
Age = 25

CatBoost:

i) Automatically converts “Mumbai” and “Electronics” into useful numeric patterns.
ii) Builds boosting trees to learn buying behavior.
iii) Combines results for final prediction.

Light GBM:

LightGBM is a gradient boosting algorithm that builds decision trees faster and more efficiently by growing trees leaf-wise instead of level-wise. It deals with nodes that have more impurity.

Ex:
Predicting whether a user will click an ad.

i) First tree makes basic predictions.
ii) Next tree focuses on users with highest error (wrong predictions).
iii) Instead of evenly growing all branches, LightGBM focuses on the most “important mistake."
iv) Final model combines all trees.

K-Nearest Neighbors (KNN):

K-Nearest Neighbors (KNN) is a simple, instance-based machine learning algorithm used for classification and regression. It makes predictions by looking at the K closest data points in the dataset.

Ex:
Suppose we want to classify whether a fruit is Apple or Orange.

We have labeled data points:

Apples are clustered in one area
Oranges in another

Now a new fruit comes in:

Step 1: Choose K = 3
Step 2: Find 3 closest fruits
2 are Apples
1 is Orange
Step 3: Majority vote

Prediction = Apple

Support Vector Machine: 

SVM is a supervised machine learning algorithm used for classification and regression but it is most commonly used for classification.

SVM finds the best boundary (hyperplane) that separates different classes with the maximum margin between them.

Support Vectors: Data points that are closest to the decision boundary and influence its position.

Ex:
two classes:
Apples
Oranges

SVM:
i) Tries many possible separating lines.
ii) Picks the one that leaves the widest gap between apples and oranges.
iii) The points closest to the line are the support vectors.
